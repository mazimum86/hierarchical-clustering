**Importinng the Libraries**

```python
import numpy as np
import pandas as pd
import seaborn as sns
from sklearn.datasets import make_blobs
import matplotlib.pyplot as plt
import warnings
warnings.filterwarnings('ignore')
```

**Creating an artificial dataset**

```python
example_dataset = make_blobs(n_samples=300, n_features=2, centers=3, cluster_std=1.3,random_state=42)
```

```python
example_dataset
```

```python
dots = example_dataset[0]
dots
```

```python
import scipy.cluster.hierarchy as sch
dendrogram = sch.dendrogram(sch.linkage(dots,method='ward'))
```

```python
plt.figure(figsize=(10,7))
plt.scatter(dots[:,0],dots[:,1]);
```

```python
from sklearn.cluster import AgglomerativeClustering
hierarchical_clustering = AgglomerativeClustering(n_clusters=3,linkage='ward')
y_hier_clus=hierarchical_clustering.fit_predict(dots)
```

```python
y_hier_clus
```

```python
plt.figure(figsize=(10,7))
for i in np.unique(y_hier_clus):
    plt.scatter(dots[y_hier_clus==i,0],dots[y_hier_clus==i,1],s=150,label=i)
plt.legend();
```
